```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# Load silver layer tables
orders = spark.table("novacart_dev.silver.slv_orders")
```

```python
# Dimension: Orders (order header information)
dim_orders = orders \
    .select(
        F.col("order_id").alias("order_key"),
        "customer_id",
        "order_date",
        "order_status",
        "order_status_clean",
        "channel",
        "country_code",
        "total_amount",
        "currency",
        "load_timestamp"
    ) \
    .distinct()

print(f"dim_orders: {dim_orders.count():,} orders")
display(dim_orders.limit(5))
```

```python
# Write dim_orders to Gold layer
dim_orders.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("novacart_dev.gold.dim_orders")

print("✓ dim_orders written to novacart_dev.gold.dim_orders")
```
