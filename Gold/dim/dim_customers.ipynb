```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

customers = spark.table("novacart_dev.silver.slv_customers")
```

```python
dim_customers = customers \
    .select(
        F.col("customer_id").alias("customer_key"),
        "registration_date",
        F.year("registration_date").alias("registration_year"),
        F.month("registration_date").alias("registration_month"),
        "load_timestamp"
    ) \
    .distinct()

display(dim_customers.limit(5))
```

```python
# Write dim_customers to Gold layer
dim_customers.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("novacart_dev.gold.dim_customers")

print("✓ dim_customers written to novacart_dev.gold.dim_customers")
```
