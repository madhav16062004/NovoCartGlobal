```python
# Load kpi tables
total_revenue_df = spark.table("novacart_dev.gold.kpi_total_revenue")
revenue_by_country = spark.table("novacart_dev.gold.kpi_revenue_by_country")
revenue_by_channel = spark.table("novacart_dev.gold.kpi_revenue_by_channel")
revenue_by_currency = spark.table("novacart_dev.gold.kpi_revenue_by_currency")
completed_order_count = spark.table("novacart_dev.gold.kpi_completed_order_count")
completed_order_rate = spark.table("novacart_dev.gold.kpi_completed_order_rate")
aov = spark.table("novacart_dev.gold.kpi_average_order_value")
top_products = spark.table("novacart_dev.gold.kpi_top_products")
active_customer_count = spark.table("novacart_dev.gold.kpi_active_customers_count")
customer_acquisition = spark.table("novacart_dev.gold.kpi_customer_acquisition")
data_quality_score = spark.table("novacart_dev.gold.kpi_data_quality_score")
dq_breakdown = spark.table("novacart_dev.gold.kpi_dq_breakdown")

print("Loaded all the kpi tables")
```

```python
from pyspark.sql.functions import round as spark_round, format_number

print("=" * 80)
print("1. TOTAL REVENUE")
print("=" * 80)

total_revenue_formatted = total_revenue_df.select(
    format_number("total_revenue_usd", 2).alias("total_revenue_usd")
)
display(total_revenue_formatted)
```

```python
print("=" * 80)
print("2. REVENUE BY COUNTRY")
print("=" * 80)

revenue_by_country_formatted = revenue_by_country.select(
    "order_country",
    format_number("revenue_usd", 2).alias("revenue_usd"),
    "order_items"
)
display(revenue_by_country_formatted)
```

```python
print("=" * 80)
print("3. REVENUE BY CHANNEL")
print("=" * 80)

revenue_by_channel_formatted = revenue_by_channel.select(
    "channel",
    format_number("revenue_usd", 2).alias("revenue_usd"),
    "order_items",
    "orders"
)
display(revenue_by_channel_formatted)
```

```python
print("=" * 80)
print("3b. REVENUE BY PRODUCT CURRENCY")
print("=" * 80)

revenue_by_currency = spark.table("novacart_dev.gold.kpi_revenue_by_currency")

revenue_by_currency_formatted = revenue_by_currency.select(
    "product_currency",
    format_number("total_local_currency", 2).alias("total_local_currency"),
    format_number("total_usd", 2).alias("total_usd"),
    "order_items"
)
display(revenue_by_currency_formatted)
```

```python
print("=" * 80)
print("4. COMPLETED ORDER COUNT")
print("=" * 80)

display(completed_order_count)
```

```python
print("=" * 80)
print("5. COMPLETED ORDER RATE")
print("=" * 80)

completed_order_rate_formatted = completed_order_rate.select(
    "total_orders",
    "completed_orders",
    format_number("completed_order_rate", 2).alias("completed_order_rate_%")
)
display(completed_order_rate_formatted)
```

```python
print("=" * 80)
print("6. AVERAGE ORDER VALUE (AOV)")
print("=" * 80)

aov_formatted = aov.select(
    format_number("avg_order_value", 2).alias("avg_order_value"),
    format_number("min_order_value", 2).alias("min_order_value"),
    format_number("max_order_value", 2).alias("max_order_value")
)
display(aov_formatted)
```

```python
print("=" * 80)
print("7. TOP PRODUCTS")
print("=" * 80)

top_products_formatted = top_products.select(
    "product_key",
    "product_name",
    "category",
    format_number("total_revenue", 2).alias("total_revenue"),
    "total_quantity_sold",
    "order_count"
)
display(top_products_formatted)
```

```python
print("=" * 80)
print("8. ACTIVE CUSTOMER COUNT")
print("=" * 80)

active_customer_count_formatted = active_customer_count.select(
    "active_customers",
    "total_customers",
    format_number("active_rate", 2).alias("active_rate_%")
)
display(active_customer_count_formatted)
```

```python
print("=" * 80)
print("9. CUSTOMER ACQUISITION")
print("=" * 80)

display(customer_acquisition)
```

```python
print("=" * 80)
print("10. DATA QUALITY SCORE")
print("=" * 80)

data_quality_score_formatted = data_quality_score.select(
    "total_records",
    "clean_records",
    "records_with_issues",
    format_number("data_quality_score", 2).alias("data_quality_score_%"),
    "total_dq_checks"
)
display(data_quality_score_formatted)
```

```python
print("=" * 80)
print("11. DATA QUALITY ISSUES - DETAILED BREAKDOWN")
print("=" * 80)

dq_breakdown = spark.table("novacart_dev.gold.kpi_dq_breakdown")
display(dq_breakdown)
```
