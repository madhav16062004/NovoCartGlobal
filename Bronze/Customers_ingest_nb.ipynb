Injecting Raw Customer Data

```python
# Read CSV from S3 into a Spark DataFrame
bronze_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("s3://databricks-project-jman/NovoCartGlobal/customers.csv")

catalog_name = "novocraftglobal"  # Change to your catalog
schema_name = "default"  # Change to your schema
table_name = "customer"
 
# Write DataFrame to Unity Catalog table
bronze_df.write.mode("overwrite").saveAsTable(f"{catalog_name}.{schema_name}.{table_name}")
 
print(f"Table created: {catalog_name}.{schema_name}.{table_name}")
```
